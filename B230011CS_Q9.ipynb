```python
def extract_policy(V, grid_size=3, gamma=0.9):
    actions = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    action_names = ['↑', '↓', '←', '→']
    policy = np.full((grid_size, grid_size), ' ', dtype=str)
    goal, pit = (2, 2), (1, 1)
    for r in range(grid_size):
        for c in range(grid_size):
            if (r, c) == goal:
                policy[r, c] = '★'
                continue
            if (r, c) == pit:
                policy[r, c] = 'X'
                continue
            best_val = -float('inf')
            best_action = ''
            for i, (dr, dc) in enumerate(actions):
                nr, nc = r + dr, c + dc
                if 0 <= nr < grid_size and 0 <= nc < grid_size:
                    val = (10 if (nr, nc) == goal else (-10 if (nr, nc) == pit else -1)) + gamma * V[nr, nc]
                    if val > best_val:
                        best_val = val
                        best_action = action_names[i]
            policy[r, c] = best_action
    return policy
print(extract_policy(optimal_values))
```
